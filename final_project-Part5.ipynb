This dataset includes scores from three test scores of students at a public school and a variety of personal and socio-economic factors that may have interaction effects upon them.

##### Limitations:

We will note here few limitations we encounterd with:
1. We dont know how the data was gathered and because of that there may be some biases in the data; for example: a measuring bias that could affect our results is that a student that registered for a preperation course but didnt complete it, it could affect the correlation between the features 'TestPrep' and 'MathScore' because there could be students that we thought they completed the course, they got high grade, but they didnt really complete the course. The potential Consequences is how trust worthy our classifier in part 4.

2. For handling the missing data we have several options:
- Removing the observations with missing data. This is the simplest way, but it reduces the number of observations by nearly 37%.It also might cause some biases with the data.
- Replace each missing observation with the mean or median. This is not relevant becauses all the missing data is categorial features.
- Replace each missing observation with a random value of the feature. We think this is not a good idea since we have shown  there is a correlation between each feature that has missing values and other features. For example with the Math exam score.
- Replace  each missing observation with the most frequent value of each feature. From the same reason of the previous option this is not a good way for our opinion.
- We can also remove a column with missing data(the feature).It requiers us to check the correlation between the variable we want to classiy in part 4 and the rest of the features(we also need to split to train and test and use dummis).It is possible but has several consequences, like choosing right now the variable we seek to classify(we might choose it later), and it is too complex.
- Removing the column with the most missing records and also remobing missing observations as we mentioned in the first option.It has some disatvantages like we discussed earlier.
We chose to remove the observations with missing data, we are aware that there may be potential biases, but all the options have Disadvantages, we thought that its the reasonable way of all.
